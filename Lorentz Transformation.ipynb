```python
#Lorentz Transformation

#length Contraction
beta=0.6 
gam=1/np.sqrt(1-beta**2)
lt=gam*np.array([[1,-beta],[-beta,1]])
ilt=np.linalg.inv(lt)
murph=np.array([[6],[10]])
cooper=lt@murph
cooper
```

```python
#Time Dilation
beta=0.6 
gam=1/np.sqrt(1-beta**2)
lt=gam*np.array([[1,-beta],[-beta,1]])
ilt=np.linalg.inv(lt)
murph=np.array([[10],[0]])
cooper=lt@murph
cooper
```

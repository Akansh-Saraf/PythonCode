```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
df=pd.read_csv(r'ViewingActivity.csv')

df=df.drop(["Device Type","Bookmark","Latest Bookmark","Country","Supplemental Video Type"],axis=1)
df['Start Time'] = pd.to_datetime(df['Start Time'])
df['Attributes']=df["Attributes"].astype("string")
df['Duration'] = pd.to_timedelta(df['Duration'])
df=df.fillna("0")
df = df[df['Attributes'].str.contains('0', regex=False)]
df=df.reset_index()
df=df.drop(["index"],axis=1)
delta=pd.Timedelta(days=0, hours=0, minutes=2, seconds=0)
df=df[df['Duration']>delta]
df
```

```python
kuku=df[df['Profile Name'].str.contains('kuku', regex=False)]
delta=pd.Timedelta(days=0, hours=0, minutes=2, seconds=0)
kuku=kuku[kuku["Duration"]>delta]
kuku=kuku.sort_values(by="Start Time", ascending=False)
kuku.head(50)
s=kuku['Title'].str.contains('Seinfeld', regex=False)
kuku.loc[kuku['Title'].str.contains('Seinfeld', regex=False), 'Title 1'] = 'Seinfeld'
kuku.loc[kuku['Title'].str.contains('The Office', regex=False), 'Title 1'] = 'The Office'
kuku.loc[kuku['Title'].str.contains('Friends', regex=False), 'Title 1'] = 'Friends'
kuku.loc[kuku['Title'].str.contains('Community', regex=False), 'Title 1'] = 'Community'
pivot=kuku.pivot_table(index=kuku.index,columns=['Title 1'], values='Duration')
pivot.sum()
```

```python
seinfeld=kuku[kuku['Title 1']=='Seinfeld']
seinfeld.pivot_table(index=seinfeld.index,values=['Duration','Start Time'],columns="Title 1")
```

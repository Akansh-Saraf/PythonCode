```python
import numpy as np
import math
from sympy import *
import sympy as sy
x,y,z,t,w,A=symbols('x y z t w A',real=True)
f,g=symbols('f g',cls=Function)
```

```python
# Quadratic Equation
exp=x**2+5*x+6
eq=Eq(exp,0)
solve(eq)
```

```python
# Simultaneous Linear Eq
eq1=x+y-3
eq2=y+z-5
eq3=x+y+z-6
eqs=[eq1,eq2,eq3]
solve(eqs)
```

```python
# Simultaneous Non Linear Eq
eq1=x**2-y
eq2=y-x
solve([eq1,eq2])
```

```python
# Calculus
a=sin(x)+x**3
print(a.diff(x))
print(integrate(a,x))
integrate(sy.exp(-x**2),(x, 0, oo))
```

```python
# Differential Equations
fp=diff(f(t),t)
fpp=f(t).diff(t,2)
diffeq=Eq(f(t).diff(t,t),-w**2*f(t))
dsolve(diffeq,f(t),ics={f(0):A,fp.subs(t,0):0})
```

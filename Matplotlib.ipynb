```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10,1)
plt.plot(x)
```

```python
y1=np.zeros(10)
y2=np.arange(0,10,1)
def f(x):
    return x**2

plt.title("Graph Title")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.grid()

plt.plot(x,y1,x,y2,x,f(x),marker='.')
plt.show()
```
